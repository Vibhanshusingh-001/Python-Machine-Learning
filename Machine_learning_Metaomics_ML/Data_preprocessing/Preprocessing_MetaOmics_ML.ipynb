FOR ALPHA **DIVERSITY**:

**Alpha diversity = diversity within a single sample (richness + evenness).**

**Beta diversity = diversity between samples (community composition differences).**

````python
import pandas as pd
import numpy as np

def chao1(counts):
    """Compute Chao1 index for one sample (counts = array of species counts)."""
    counts = np.array(counts)
    S_obs = np.sum(counts > 0)         # observed species richness
    F1 = np.sum(counts == 1)           # singletons
    F2 = np.sum(counts == 2)           # doubletons

    if F2 == 0:  # avoid division by zero
        return S_obs + (F1 * (F1 - 1)) / 2 if F1 > 1 else S_obs
    else:
        return S_obs + (F1**2) / (2 * F2)

# Apply Chao1 to each sample (column)
chao1_values = brack_df.apply(chao1, axis=0)

# Put into a DataFrame
chao1_df = pd.DataFrame({"Sample": chao1_values.index, "Chao1": chao1_values.values})

# Show first results

chao1_df
````

# Chao1 formula

computing the Chao1 index, which is a widely used alpha diversity metric in ecology and metagenomics. It estimates the true species richness of a community by correcting for unobserved (rare) species using the number of singletons and doubletons.



F1 â†’ number of species that appear only once (singletons).

F2 â†’ number of species that appear exactly twice (doubletons).

# ðŸ”¹ Spearman Correlation (Ï or *Spearmanâ€™s Rho*)

* It measures how well the **relationship between two variables** can be described using a **monotonic function** (values move together in a consistent direction, not necessarily linear).
* Instead of comparing raw values like Pearson correlation does, Spearman uses **ranks**:

  * Example: If species A abundance ranks higher in Disease samples than in Healthy samples, Spearman correlation will be positive.
  * If species A abundance tends to be lower in Disease samples, correlation will be negative.
* **Range:** -1 to +1

  * **+1** â†’ perfect positive monotonic relationship
  * **-1** â†’ perfect negative monotonic relationship
  * **0** â†’ no monotonic relationship

ðŸ‘‰ Why  here? Because microbiome data (species abundances) are often **non-normally distributed** and contain outliers. Spearman is **non-parametric** (no assumption of linearity or normality).

---

## ðŸ”¹ p-value

* The **p-value** tells you whether the observed correlation is likely to have happened **by chance** under the null hypothesis (*no association*).
* Small p-value (e.g. < 0.05) â†’ the correlation is **statistically significant**.
* Large p-value â†’ the observed correlation may be just random noise.

Formally:

* **Null hypothesis (Hâ‚€):** Spearman correlation = 0 (no association)
* **Alternative hypothesis (Hâ‚):** Spearman correlation â‰  0 (some association)

---

## ðŸ”¹ Example

* Suppose `Species_X` has **Ï = 0.6, p = 0.001** with phenotype "Disease":

  * Ï = 0.6 â†’ strong positive correlation (Species\_X is more abundant in Disease).
  * p = 0.001 â†’ very unlikely to be random, so itâ€™s a meaningful association.

* If `Species_Y` has **Ï = -0.4, p = 0.2**:

  * Ï = -0.4 â†’ moderate negative correlation.
  * p = 0.2 â†’ too high â†’ not statistically significant, so we canâ€™t trust it.

---

âœ…

* **Spearman correlation** tells you the *strength and direction* of association.
* **p-value** tells you if that association is likely real or just random.

# Centered Log-Ratio (CLR) transformation

---

## ðŸ”¹ The Problem: Compositional Data

* Microbiome data (counts, relative abundances) are **compositional**.

  * Example: if a sample has 3 species with proportions `[0.2, 0.3, 0.5]`, they must sum to 1.
* This constraint causes problems:

  * If one species increases, at least one other must decrease, even if biologically independent.
  * Correlations and distance measures can become misleading.

---

## ðŸ”¹ The Idea of CLR

The CLR transform fixes this by turning each component into a **log-ratio relative to the geometric mean** of the sample.

### Formula

For a sample with values:

$$
x = [x_1, x_2, ..., x_D]
$$

The **geometric mean** is:

$$
g(x) = \left(\prod_{i=1}^D x_i \right)^{1/D}
$$

The **CLR-transformed vector** is:

$$
CLR(x) = \Big[ \ln\frac{x_1}{g(x)}, \ln\frac{x_2}{g(x)}, ..., \ln\frac{x_D}{g(x)} \Big]
$$

---

## ðŸ”¹ Key Properties

1. **Centered** â†’ The transformed values sum to 0.

   $$
   \sum_{i=1}^D CLR(x_i) = 0
   $$

   (This avoids redundancy and makes data â€œbalanced.â€)

2. **Relative, not absolute** â†’ Each value is expressed as a ratio to the sampleâ€™s central tendency (geometric mean).

3. **Log-ratio scale** â†’ Handles multiplicative relationships better, and distances between samples become meaningful.

---

## ðŸ”¹ Example

Say you have a sample with abundances:

```
[2, 6, 12]
```

* Geometric mean:

  $$
  g(x) = (2 \cdot 6 \cdot 12)^{1/3} = (144)^{1/3} â‰ˆ 5.24
  $$

* CLR values:

  $$
  [\ln(2/5.24), \ln(6/5.24), \ln(12/5.24)]
  $$

  \= `[-0.96, 0.14, 0.82]`

Notice they sum to â‰ˆ 0.

---

## ðŸ”¹ Why CLR is used in microbiome analysis

* Removes the **sum-to-one constraint** of compositional data.
* Allows **Euclidean distance, PCA, regression, clustering** to be applied properly.
* Gives interpretable results: values > 0 mean "more than average abundance," values < 0 mean "less than average abundance" in that sample.

---

âœ…
CLR transformation converts compositional microbiome data into a log-ratio form where values are expressed relative to the geometric mean, fixing the compositional bias and making statistical analysis valid.



````python
import pandas as pd
import numpy as np
from scipy.stats import spearmanr
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Pivot to wide format
species_wide = new_df.pivot_table(index="Run_ID", columns="species", values="RA", fill_value=0)

# Add phenotype info
meta = new_df.drop_duplicates("Run_ID").set_index("Run_ID")[["Phenotype"]]
species_wide = species_wide.join(meta)

# 2. One-hot encode phenotypes (dummy vars)
phenotype_dummies = pd.get_dummies(species_wide["Phenotype"])
species_wide = species_wide.join(phenotype_dummies)

# 3. Spearman correlation: each species vs each phenotype dummy
meta_vars = phenotype_dummies.columns.tolist()
cor_results = []
for sp in species_wide.columns.difference(["Phenotype"] + meta_vars):
    for var in meta_vars:
        corr, pval = spearmanr(species_wide[sp], species_wide[var], nan_policy="omit")
        cor_results.append([sp, var, corr, pval])

cor_df = pd.DataFrame(cor_results, columns=["Species", "Phenotype", "SpearmanR", "pval"])

# 4. Select top 50 species by absolute correlation with any phenotype
top_species = (
    cor_df.groupby("Species")["SpearmanR"]
    .apply(lambda x: x.abs().max())
    .sort_values(ascending=False)
    .head(50)
    .index
)

heatmap_df = cor_df.pivot(index="Species", columns="Phenotype", values="SpearmanR").loc[top_species]

# 5. Heatmap
plt.figure(figsize=(8, 14))
sns.heatmap(heatmap_df, cmap="vlag", center=0, annot=True, fmt=".2f",
            cbar_kws={"label": "Spearman R"})
plt.title("Top 50 Species Correlated with Phenotypes")
plt.ylabel("Species")
plt.xlabel("Phenotype")
plt.tight_layout()
plt.show()


# Dark red / positive values â†’ the species is more abundant when that phenotype is present.

# Dark blue / negative values â†’ the species is less abundant when that phenotype is present.

# Annotated numbers = exact Spearman correlation coefficients.

# Range: +1 = perfect positive, -1 = perfect negative, 0 = no association.
````

EXPLNATION OF ABOVE CODE
---

### **Step 1. Pivot to wide format**

* Create a **samples Ã— species** matrix: rows = `Run_ID`, columns = `species`, values = relative abundance (`RA`).
* Missing values filled with `0`.
* Join phenotype info (`Phenotype`) for each sample.

---

### **Step 2. One-hot encode phenotypes**

* Convert `Phenotype` column into **dummy variables** (binary 0/1 columns, one per phenotype).
* Add these dummy columns to the wide matrix.

---

### **Step 3. Compute Spearman correlations**

* For each species column (predictor) vs each phenotype dummy column (target):

  * Compute **Spearman correlation coefficient (Ï)** and **p-value**.
* Store results in a table (`cor_df`) with columns:

  * `Species`, `Phenotype`, `SpearmanR`, `pval`.

---

### **Step 4. Select top species**

* For each species, take the **maximum absolute correlation** across all phenotypes.
* Rank species by this value, select **top 50**.
* Reshape results into a matrix (`heatmap_df`) of species Ã— phenotype, values = Spearman R.

---

### **Step 5. Plot heatmap**

* Draw a heatmap of correlations:

  * Rows = top 50 species.
  * Columns = phenotypes.
  * Color = correlation strength (blueâ€“red, centered at 0).
  * Annotated with correlation values.
* Add title, labels, and adjust layout for readability.

---



````python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming new_df is the merged DataFrame
# Group by Health_status and species, summing the RA (relative abundance)
#aggregated_df = new_df.groupby(['Health_status', 'species'])['RA'].sum().unstack().fillna(0)


aggregated_df = new_df.groupby(['Health_status', 'species'])['RA'].mean().unstack().fillna(0)

# Select top N species by total abundance for clarity (e.g., top 10 species)
top_n = 10
top_species = aggregated_df.sum(axis=0).nlargest(top_n).index
plot_df = aggregated_df[top_species]

# Create a stacked bar plot
plot_df.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='tab20')

# Customize the plot
plt.title('Species Abundance by Health Status')
plt.xlabel('Health Status')
plt.ylabel('Relative Abundance (RA)')
plt.legend(title='Species', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()


````

# **DIFFERNCE** BETWEEN **TRANSFORMATION** & **NORMALISATION**



## ðŸ”¹ **Normalization**

* **Goal**: Brings all values into a **fixed range**, usually \[0, 1] or \[-1, 1].
* **Formula (Minâ€“Max Scaling)**:

  $$
  X' = \frac{X - X_{min}}{X_{max} - X_{min}}
  $$
* **Effect**: Only changes the **scale** of the data, without changing its shape or distribution.
* **When used**:

  * Distance-based algorithms (KNN, K-means, Neural Networks)
  * Ensuring features with different units (e.g., height in cm, weight in kg) contribute equally

ðŸ‘‰ Example: If we normalize `[10, 20, 30]` â†’ `[0, 0.5, 1]`

---

## ðŸ”¹ **Transformation**

* **Goal**: Changes the **distribution or shape** of data (not just scale).

* **Types of transformations**:

  * **Log Transformation** â†’ reduces skewness, compresses large values
  * **Square Root / Cube Root Transformation** â†’ reduces effect of large outliers
  * **Z-score Standardization (Scaling)**:

    $$
    X' = \frac{X - \mu}{\sigma}
    $$

    Makes mean = 0, standard deviation = 1
  * **Box-Cox, Yeo-Johnson** â†’ more advanced transformations for normality

* **Effect**:

  * Can make skewed data more **normally distributed**
  * Reduces impact of **outliers**
  * Makes statistical tests and ML models (like regression, PCA) work better

ðŸ‘‰ Example: Log-transform `[1, 10, 100, 1000]` â†’ `[0, 1, 2, 3]` (distribution becomes tighter).

---

##  Key Difference

* **Normalization** = rescaling values to a bounded range (keeps shape).
* **Transformation** = modifying data distribution or properties (changes shape).

---





| Aspect                 | **Normalization**                                         | **Transformation**                                               |
| ---------------------- | --------------------------------------------------------- | ---------------------------------------------------------------- |
| **Definition**         | Rescales values into a fixed range (e.g., \[0,1])         | Changes the distribution or shape of data                        |
| **Main Goal**          | To put features on the same **scale**                     | To reduce skewness, stabilize variance, or approximate normality |
| **Formula (example)**  | Minâ€“Max Scaling:  $\frac{X - X_{min}}{X_{max} - X_{min}}$ | Log, Square Root, Z-score, Box-Cox, Yeo-Johnson etc.             |
| **Effect on Shape**    | Keeps the shape of distribution **same**                  | Modifies the distribution (can make skewed â†’ normal)             |
| **Effect on Outliers** | Outliers stay (but scaled)                                | Outliersâ€™ impact reduced (e.g., log compresses large values)     |
| **Range of Output**    | Usually \[0,1] or \[-1,1]                                 | Not fixed, depends on method                                     |
| **When Used**          | KNN, K-means, Neural Networks, distance-based models      | Regression, PCA, statistical tests, skewed data correction       |
| **Example Input**      | \[10, 20, 30]                                             | \[1, 10, 100, 1000]                                              |
| **Example Output**     | Normalized â†’ \[0, 0.5, 1]                                 | Log-transformed â†’ \[0, 1, 2, 3]                                  |

---

ðŸ‘‰ In short:

* **Normalization** = scaling only.
* **Transformation** = scaling **+ distribution shaping**.
